```python
movies=df.drop(['Poster_Url','Overview'],axis=1)
```

```python
movies
```

```python
#What is most of the film genre?
movies_data={}
for genre in movies['Genre']:
    for i in genre.split(', '):
        if i not in movies_data:
            movies_data[i]=1
        else:
            movies_data[i]=movies_data[i]+1
```

```python
movies_data_df=pd.DataFrame(list(movies_data.items()),columns=['Genre','number of movies'])
movies_data_df
```

```python
#What is most of the film genre?
fig= plt.subplots(figsize=(20, 10))
sns.barplot(data=movies_data_df,x='Genre',y='number of movies')
plt.show()
```
